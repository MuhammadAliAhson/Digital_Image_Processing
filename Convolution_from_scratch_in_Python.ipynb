```python
import numpy as np
import pandas as pd
```

```python
x = np.array([[1,2,3],[1,2,4],[0,5,12]])
f = np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
```

```python
def convolution(input_image, kernel, padding = 0):
    if padding == 0:
        input_size_x = input_image.shape[0]
        input_size_y = input_image.shape[1]
        kernel_size = kernel.shape[0]
        output_image = np.empty((input_size_x - kernel_size + 1,input_size_y - kernel_size + 1),dtype = int)
        for i in range(input_size_x - kernel_size + 1):
            for j in range(input_size_y - kernel_size + 1):
                h = 0
                for k in range(i,i+kernel_size):
                    for l in range(j,j+kernel_size):
                        h += input_image[i+k][j+l] * kernel[k][l]
                output_image[i,j] = h
    else:
        
    return output_image
```

```python
a = convolution(x,f,0)
```
